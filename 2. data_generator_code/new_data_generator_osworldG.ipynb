```python
from PIL import Image, ImageDraw
import os
import io
import pandas as pd
import json

## download from 'https://github.com/xlang-ai/OSWorld-G/tree/main/benchmark'

root_path = "/Data/OSWorld_G/"

json_path = root_path + "OSWorld-G.json"
with open(json_path, 'r', encoding='utf-8') as file:
    data_info = json.load(file)

path = root_path + "os_world_G_noImage.parquet"
df = pd.read_parquet(path)
idx_list = list(df.index)
```

```python
df['image_bytes'] = None  
for idx in idx_list:    
    problem = df.loc[idx]
    image_path = root_path + "images/" + problem['img_filename']
    source_image = Image.open(image_path)

    byte_data = io.BytesIO()
    source_image.save(byte_data, format='PNG')
    source_image_bytes = byte_data.getvalue()
    df.at[idx, "image_bytes"] = source_image_bytes
```
